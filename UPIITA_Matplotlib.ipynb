### Nuestra primer gráfica :D

Las figuras de matplotlib se basan en crear una figura y dibujar sobre ella hasta que haya un cambio de figura y se decida dibujar en otra o mostrar la que se creó primero

```python
x = [i for i in range(20)]
y = [i**2 for i in range(20)]
```

```python
Figura = plt.figure() # Consejo personal, siempre guarden la información de la figura en una variable
plt.plot(x, y)
plt.show() # No pasa nada si no lo ponen, pero es una buena práctica
```

#### Para cambiar la "borrosidad"

```python
Figura = plt.figure(dpi=150)
plt.plot(x, y)
plt.show()
```

#### Para cambiar el tamaño

```python
Figura = plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.show()
```

#### Gráfica scatter + línea

```python
x = np.random.randint(0, 10, size=(10))
y = np.random.randint(0, 10, size=(10))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

Figura = plt.figure(figsize=(6,6))

plt.plot(x, y, '--o')

# - : Línea continua
# -- : Línea de guiones
# -.: Línea de guión punto
# : : Línea de puntos

# Se combinan con marcadores
# o : Punto
# * : Estrella
# ^ : Triángulo
# v : Triángulo x2

plt.title('Otra gráfica bien chida :3')
plt.ylabel('Aquí va el eje y')
plt.xlabel('Aquí va el eje x')
plt.grid(alpha=0.5)
plt.show()
```

### Diagramas de Caja

```python
A = [0.5 for i in range(10)]
B = [1.5 for i in range(12)]
C = [2.5 for i in range(12)]
D = [3.5 for i in range(10)]
E = [4.5 for i in range(7)]

data = A+B+C+D+E

fig = plt.figure(figsize =(10, 7))

plt.boxplot(data)
plt.title('Diagrama de Cajitas')
plt.show()
```

```python
A = [0.5 for i in range(10)]
B = [1.5 for i in range(12)]
C = [2.5 for i in range(12)]
D = [3.5 for i in range(10)]
E = [4.5 for i in range(7)]

data = np.array(A+B+C+D+E)

fig = plt.figure(figsize =(10, 7))

plt.boxplot(data, notch=True, vert=False, widths=0.5, labels=['Una medida'])
plt.title('Diagrama de Cajitas')
plt.show()
```

```python
A = [0.5 for i in range(10)]
B = [1.5 for i in range(12)]
C = [2.5 for i in range(12)]
D = [3.5 for i in range(10)]
E = [4.5 for i in range(7)]

data = np.array(A+B+C+D+E)

fig = plt.figure(figsize =(10, 7))

plt.boxplot([data, data*2, data/2], notch=True, vert=False, widths=0.5, labels=['Una medida', 'Otra medida', 'Ultima'])
plt.title('Diagrama de Cajitas')
plt.show()
```

```python
A = [0.5 for i in range(10)]
B = [1.5 for i in range(12)]
C = [2.5 for i in range(12)]
D = [3.5 for i in range(10)]
E = [4.5 for i in range(7)]

data = np.array(A+B+C+D+E)

fig = plt.figure(figsize =(10, 7))

Cajas = plt.boxplot([data, data*2, data/2], notch=True, vert=False, widths=0.5, labels=['Una medida', 'Otra medida', 'Ultima'])

Cajas['boxes'][0].set(color='#FF0000', linewidth=5)
Cajas['boxes'][1].set(color='#00FF00', linewidth=5)
Cajas['boxes'][2].set(color='#0000FF', linewidth=5)

plt.title('Diagrama de Cajitas')
plt.grid(alpha=0.4)

plt.show()
```
